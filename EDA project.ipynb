```python
directors = df_Moviesonly['director'].astype(str).str.split(',').explode().str.strip()
directors = directors[directors != 'Unavailable']
director_counts = directors.value_counts()

plt.figure(figsize=(16,6))
sns.barplot(x=director_counts.head(10).values, y=director_counts.head(10).index, palette="viridis")
plt.title('Top 10 Directors with the Most Contents on Disney+')
plt.xlabel('Number of Contents')
plt.ylabel('Director')
plt.show()
```

```python
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(16,6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Distribution of Contents by Rating on Disney+')
plt.xlabel('Rating')
plt.ylabel('Number of Contents')
plt.xticks(rotation=30)
plt.show()
```

```python
country_data = df['country'].str.split(',').explode()
country_data = country_data[country_data != 'Unavailable']
country_counts = country_data.str.strip().value_counts()
plt.figure(figsize=(20,6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('All Countries by Number of Contents on Disney+')
plt.xlabel('Country')
plt.ylabel('Number of Contents')
plt.xticks(rotation=65)
plt.show()
```
